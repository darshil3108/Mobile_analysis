```python
import numpy as np
import pandas as pd

mobiles=pd.read_csv("mobiles.csv")
mobiles= mobiles.drop(mobiles.columns[0], axis=1)

print(mobiles)
mobiles.isna().sum()

mobiles[['Width', 'Height', 'Depth', 'Weight']] = mobiles[['Width', 'Height', 'Depth', 'Weight']].replace(r'[^\d.]+', '', regex=True).astype(float)
mobiles['Weight'] = mobiles['Weight'].round(2)

col=[17,20,22,26,28]
mobiles.drop(mobiles.columns[col], axis=1, inplace=True)

mobiles.drop(mobiles.columns[7], axis=1, inplace=True)

mobiles['Smartphone'] = mobiles['Smartphone'].str.strip().str.lower()

no_smartphone=mobiles[mobiles['Smartphone'] == 'no']
no_smartphone.iloc[:, 62] = no_smartphone.iloc[:, 62].str.extract('(\d+)').astype(int)

smartphones=mobiles[mobiles['Smartphone'] == 'yes']

```
